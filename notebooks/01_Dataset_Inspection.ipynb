Data Cleaning & Preprocessing (Part 1)
🎯 Goal

By the end of this phase, we want to transform our raw dataset into a clean dataset that is ready for EDA and machine learning.

```python
#Step 1 — Convert Blank Strings to Missing Values

#Whenever you see a blank string, consider it a missing value.

import numpy as np

df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)
```

```python
#Step 2 — Verify the Conversion
df["TotalCharges"].isnull().sum()
```

```python
#Step 3 — Convert the Datatype
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

*Why pd.to_numeric() instead of astype(float)?*

Good question.

pd.to_numeric() is more robust and commonly used during preprocessing. It also provides options (like errors='coerce') for handling invalid values in larger datasets.

It's a good habit to use it when converting columns that come from CSV files.

```python
#Step 4 — Verify Again
df.info()
```

```python
#Step 5 — Fill Missing Values
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
#Step 6 — Final Verification
df.info()
```

```python
df.isnull().sum()
```
